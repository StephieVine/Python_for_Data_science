```python
import matplotlib.pyplot as plt

fruits = ["Orange", "Banana", "Apple", "Mango", "Pineapple"]
quantity = [15, 10, 25, 15, 35]
explode = (0, 0, 0.1, 0, 0.1)#to bring out of the piechart #tuple value

ax1 = plt.subplots()
# plt.pie(quantity, labels=fruits, autopct='%1.1f%%',shadow=False, startangle=90)
# plt.pie(quantity, labels=fruits, explode=explode, autopct='%1.1f%%',shadow=True, startangle=90)
plt.pie(quantity, labels=fruits, explode=explode,shadow=True, startangle=90, colors=['floralwhite', 'rosybrown', 'gray', 'saddlebrown', 'springgreen'])
#startangle:all the slices are rotated counter-clockwise by 90 degrees
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

fruits = ["Orange", "Banana", "Apple", "Mango", "Pineapple"]
quantity = [15, 10, 25, 15, 35]
plt.xlabel('Fruits')
plt.ylabel("Quantity")
plt.title("Fruit market stats")

plt.bar(fruits, quantity)#x,y axis
plt.show()
#bar chart
```
