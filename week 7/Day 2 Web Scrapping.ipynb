```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

url = 'https://www.jumia.com.ng/mlp-stay-connected-deals/smartphones/'
page = requests.get(url)
soup = bs(page.content, 'html5lib') #ebautiful soup is to pass the content
# print(soup.prettify())
table = soup.find('div', attrs={'class': '-paxs row _no-g _4cl-3cm-shs'})
# print(table.prettify())
articles = table.findAll('article', attrs={'class': 'prd _fb col c-prd'}) #this finds all this occurrence
print(articles[1].prettify())
#using for loop to iterate over the items and extrac the items and convert to a file
#first define an empty list[]
# items = []
df = pd.DataFrame(columns=['name', 'discounted_price ₦', 'original_price ₦', 'discount', 'rating'])

for phone in articles:
    item = {}
    item['name'] = phone.find('h3', attrs={'class': 'name'}).text
    item['discounted_price ₦'] = phone.find('div', attrs={'class': 'prc'}).text.replace('₦ ', "")
    item['original_price ₦'] = phone.find('div', attrs={'class': 'old'}).text.replace('₦ ', "")
    item['discount'] = phone.find('div', attrs={'class': 'bdg _dsct _sm'}).text
#     item['rating'] = phone.find('div', attrs={'class': 'stars -s'})
    rating = phone.find('div', class_ ='stars _s')
    if rating is not None:
        item['rating'] = rating.contents[0]
    else:
        item['rating'] = ""
    df = df.append(item, ignore_index=True)
# df.head
df.to_csv('Jumia_phone_data.csv') #convert to csv file
print('the end')
```

```python
url = 'https://hotels.ng/hotels-in-lagos'
page = requests.get(url)
soup = bs(page.content, 'html5lib') #ebautiful soup is to pass the content
# print(soup.prettify())
table = soup.find('div', attrs={'class': "dropdown__item active"})
# print(table.prettify())
articles = table.findAll('article', attrs={'class': 'col-xs-12 col-sm-8 row listing-hotels-details-box'}) #this finds all this occurrence
# print(articles[0].prettify())
# #using for loop to iterate over the items and extrac the items and convert to a file
# #first define an empty list[]
# # items = []
df = pd.DataFrame(columns=['name', 'address', 'cost', 'price'])

for hotel in articles:
    item = {}
    item['name'] = hotel.find('h3', attrs={'class': 'name'}).text
    item['address'] = hotel.find('div', attrs={'class': 'prc'}).text
    item['cost'] = hotel.find('div', attrs={'class': 'old'}).text
    item['price'] = hotel.find('div', attrs={'class': 'bdg _dsct _sm'}).text
#     item['rating'] = hotel.find('div', attrs={'class': 'stars -s'})
    df = df.append(item, ignore_index=True)
# # df.head
df.to_csv('Hotel_data.csv') #convert to csv file
print('the end')
```
