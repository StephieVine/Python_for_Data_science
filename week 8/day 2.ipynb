```python
import pymysql

db = pymysql.connect(host='localhost', user='root', password='1234', database='univelcity', charset='utf8mb4', cursorclass=pymysql.cursors.DictCursor) #this connect method expects keyword arguement eg like host

#prepare a cursor object for access using cursor() method
cursor=db.cursor()

#prepare sql query
query = ("select * from students")

#execute the query
cursor.execute(query)

#fetchall records is a method used to fetch data from a record on the database
data = cursor.fetchall()

# print(data)
# print(data[0]) #it fetches the first record and it comes as a dictionary
#how to display the record in simpler and understanding form. Loop over the records
print('id \tname \t\t\tage \taddress \temail')
for student in data:
    id, name, age, address, email = student.values()
#     print(student.values())
#     print("{}, {}, {}, {}, {}".format(id, name, age, address, email))
    print(str(id) + "\t" + name + "\t" + str(age) + "\t" + address + "\t" + email)
```

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

url = 'https://www.jumia.com.ng/mlp-stay-connected-deals/smartphones/'
page = requests.get(url)
soup = bs(page.content, 'html5lib') #ebautiful soup is to pass the content
# print(soup.prettify())
table = soup.find('div', attrs={'class': '-paxs row _no-g _4cl-3cm-shs'})
# print(table.prettify())
articles = table.findAll('article', attrs={'class': 'prd _fb col c-prd'}) #this finds all this occurrence
# print(articles[1].prettify())
#using for loop to iterate over the items and extrac the items and convert to a file
#first define an empty list[]
# items = []
df = pd.DataFrame(columns=['name', 'discounted_price', 'original_price ₦', 'discount', 'rating'])

for phone in articles:
    item = {}
    item['name'] = phone.find('h3', attrs={'class': 'name'}).text
    discount_price = phone.find('div', attrs={'class': 'prc'})
    if discount_price is not None:
       item['discounted_price'] = discount_price.text.replace("₦ ", "")
#     item['discounted_price'] = phone.find('div', attrs={'class': 'prc'}).text.replace('₦ ', "")
    original_price = phone.find('div', attrs={'class': 'old'})
    if original_price is not None:
        item['original_price'] = original_price.text.replace("₦ ", "")
    discount = phone.find('div', attrs={'class': 'bdg_dsct _sm'})
    if discount is not None:
        item['discount'] =  discount.text
    rating = phone.find('div', class_='stars _s')
    if rating is not None:
        item['rating'] = rating.contents[0]
    else:
        item['rating'] = ""
    

# #     item['original_price'] = phone.find('div', attrs={'class': 'old'}).text.replace('₦ ', "")
# #     item['discount'] = phone.find('div', attrs={'class': 'bdg _dsct _sm'}).text
# #     item['rating'] = phone.find('div', attrs={'class': 'stars -s'})
#     rating = phone.find('div', class_ ='stars _s')
#     if rating is not None:
#         item['rating'] = rating.contents[0] #or . find_net(text=true)
#     else:
#         item['rating'] = ""
    df = df.append(item, ignore_index=True)
# # df.head
# df.to_csv('Jumia_phone_data.csv') #convert to csv file
# print('the end')

try:
    engine = create_engine("mysql://root:1234@localhost/univelcity")#to specify the connection credential
    with engine.begin() as connection:
        df.to_sql(name="web_scrapping_information", con=connection, if_exists='append', index=False) #(df means Data frame)
except Exception as e:
    print(e)
print("We are ready!")
```
